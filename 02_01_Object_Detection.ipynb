# Object Detection⚡⚡

- HAAR Cascade Classifier

- Go to this [link](https://github.com/opencv/opencv/tree/master/data/haarcascades) and download haarcascade_frontalface_default.xml. ⏬⏬

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

- To detect the face:
    1. Read Image
    2. Convert into Grayscale.
    3. Apply Cascade Classifier

```python
img = cv2.imread('data/chris.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
haar = cv2.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
faces = haar.detectMultiScale(gray,1.3,5)
```

```python
print(faces) # x,y,width,height
```

```python
x1 = faces[0][0]
y1 = faces[0][1]
x2 = faces[0][2]
y2 = faces[0][3]
```

```python
cv2.rectangle(img,(x1,y1),(x1+x2,y1+y2),(0,255,0),5)
plt.title('FACE DETECTION USING HAAR CASCADE')
plt.imshow(img)
```

```python
cv2.imshow('object_detect',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Crop the image
faceCrop = img[y1:y1+y2,x1:x1+x2]
plt.imshow(faceCrop)
```

```python
# Save the face
cv2.imwrite('./data/chris_detected_face.png',faceCrop)
```
