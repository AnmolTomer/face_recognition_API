## Pixels 
- For an image we have pixel, pixels are also known as values of the image and ranges from 0 - (2^n-1)
- For an 8 bit image, n = 8, range of pixel values is 0 - 255

![](https://i.imgur.com/4lLXDjj.png)

![](https://i.imgur.com/uvPlEZB.png)

```python
# Read the image
img = cv2.imread('./data/test.jpg')
img.shape
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
slice = gray[0:10,0:10]
plt.imshow(slice,cmap='gray')
```

## Image Resizing
- Resize is very useful technique used to structure our data. 2 kinds of resize: Shrink and Enlarge.

![](https://i.imgur.com/6t7jIwP.png)

```python
# Shrink a 2D Array
img.shape # height = 893 , width = 1000
```

```python
img_resized = cv2.resize(gray,(500,400),cv2.INTER_AREA)
img_resized.shape
```

```python
plt.imshow(img_resized,cmap='gray') # Resized image.
```

```python
# Enlarging the image say we want height = 1600 and width = 900
img_en = cv2.resize(gray,(1600,900),cv2.INTER_CUBIC)
img_en.shape
```

```python
plt.imshow(img_en,cmap='gray') # Resized image.
plt.show()
```
